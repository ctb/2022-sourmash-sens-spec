```python
all_idx = set(db.idx_to_ident)
print(len(all_idx))
```

```python
assignments = {}

xx = []

for n, hashval in enumerate(db.hashvals):
    if n % 100000 == 0:
        print('...', n, len(all_idx))
        #if n: break
    idxlist = set(db.hashval_to_idx[hashval])
    if len(idxlist) == 1: # perfectly discriminating k-mer? yes!
        all_idx -= idxlist # remove from list of idxs without such.
```
